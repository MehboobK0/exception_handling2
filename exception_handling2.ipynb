By inheriting from the Exception class, our Custom Exception inherits all the properties and behaviors of the Exception class, such as error message handling, traceback generation, and error propagation. This allows our Custom Exception to seamlessly integrate with the Python exception handling mechanism and provides consistency in exception handling across different types of exceptions.

```python
# Q2. Python Exception Hierarchy:
def print_exception_hierarchy(exception_class, depth=0):
    print(' ' * depth + exception_class.__name__)
    for subclass in exception_class.__subclasses__():
        print_exception_hierarchy(subclass, depth + 2)

print_exception_hierarchy(BaseException)
```
